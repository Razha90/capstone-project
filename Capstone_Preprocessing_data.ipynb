```python
from google.colab import auth
import pandas as pd
import gspread
from google.auth import default
import numpy as np

auth.authenticate_user()
creds, _ = default()
gc = gspread.authorize(creds)
```

```python
spreadsheet = gc.open_by_url('https://docs.google.com/spreadsheets/d/1LD3oBPCwI4BvmhbsD77ia9bcStlyioJVULuZho5Bfw0/edit?usp=sharing')
worksheet = spreadsheet.sheet1
data = worksheet.get_all_values()
```

```python
df = pd.DataFrame(data[1:], columns=data[0])
df.head()
```

## Menyesuaikan Nama Coloumn

```python
## Menganti Beberapa Nama Coloumn
df = df.rename(columns={'Nama (Opsional)': 'Nama', 'Bagaimana mood Anda hari ini?':'Mood Harian', 'Seberapa besar tingkat stres Anda hari ini?':'Tingkat Stress','Seberapa baik kualitas tidur Anda?':'Kualitas Tidur','Seberapa aktif Anda hari ini?':'Aktivitas', 'Apakah ada kejadian atau hal yang membuat Anda merasa lebih stres atau tertekan hari ini? Jika ada, ceritakan!':'Stress Activity' })

## Memisahkan beberapa coloumn
df_manfaat_aplikasi = df['Menurut Anda, seberapa bermanfaatkah aplikasi pemantauan kesehatan mental?']
df_cerita_pengalaman = df['Ceritakan pengalaman Anda menggunakan aplikasi-aplikasi untuk kesehatan mental, seperti Riliv, dan sebagainya.']
df_email = df['Email Address']
df_nama = df['Nama']

## Hapus Coloumn yang sudah dipisah
df = df.drop(columns=['Menurut Anda, seberapa bermanfaatkah aplikasi pemantauan kesehatan mental?'],axis=1)
df = df.drop(columns=['Ceritakan pengalaman Anda menggunakan aplikasi-aplikasi untuk kesehatan mental, seperti Riliv, dan sebagainya.'],axis=1)
df = df.drop(columns=['Email Address'],axis=1)
df = df.drop(columns=['Nama'],axis=1)

df.head()
```

## Cleaning Data

## Change Type Data

```python
print(df.info())
```

```python
## Ubah Menjadi Numeric beberapa type data
df['Mood Harian'] = pd.to_numeric(df['Mood Harian'])
df['Tingkat Stress'] = pd.to_numeric(df['Tingkat Stress'])
df['Kualitas Tidur'] = pd.to_numeric(df['Kualitas Tidur'])
print(df.info())
```

## Cleaning Duplicate

```python
print('Sebelum Duplicate', df.shape)
df = df.drop_duplicates(subset=['Mood Harian', 'Tingkat Stress', 'Kualitas Tidur', 'Aktivitas', 'Stress Activity'])
print('Sesudah Duplicate', df.shape)
```

# Encoding Data

```python
aktivitas_mapping = {'Pasif': 0, 'Sedang': 1, 'Aktif': 2}
df['Aktivitas'] = df['Aktivitas'].map(aktivitas_mapping)

usia_mapping = {'12-18': 0, '19-24': 1, '25-31': 2, 'Yang lain': 3}
df['Usia'] = df['Usia'].map(usia_mapping)
df['Usia'] = pd.to_numeric(df['Usia'])


df.head()
```

```python
def split_data(df):
  split_idx = int(0.8 * len(df))

  df_80 = df.iloc[:split_idx]
  df_20 = df.iloc[split_idx:]

  return df_80, df_20
```

```python
df_train, df_test = split_data(df)
print(df_train.shape)
print(df_test.shape)
```

```python
df_train.to_csv('df_train.csv', index=True)
df_test.to_csv('df_test.csv', index=True)
```
